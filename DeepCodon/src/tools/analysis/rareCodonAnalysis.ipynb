```python
import pandas as pd

from DeepCodon.src.services.appendixTool import *
# from DeepCodon.src.services.foundRareCodon import smoon
```

```python
fastaTokenPath = "data/datasets/test_data/teach/geneScript/"
othersPath = "data/datasets/test_data/cai/"
```

```python
df_deepcodon = pd.read_csv("src/tools/analysis/seqs/4deepcodon2aRarecodon.csv")
df_deepcodon.rename(columns={"outputseq": "deepcodon"}, inplace=True)


df_geneScript = pd.read_csv(
    othersPath + "genescript_val.csv",
    header=None,
    names=["id", "input", "genescript", "ic", "oc"],
)

df_icor = pd.read_csv(
    othersPath + "icor_val.csv",
    header=None,
    names=["id", "input", "icor", "ic", "oc"],
)

df_codonop = pd.read_csv(
    othersPath + "codonop_val.csv",
    header=None,
    names=["id", "input", "codonop", "ic", "oc"],
)

df_deepcodonft = pd.read_csv(
    othersPath + "mymodel_val.csv",
    header=None,
    names=["id", "input", "deepcodonft", "ic", "oc"],
)
```

```python
token = {}
for i in range(1, 11):
    df_deepcodon_token = assert_all_codon_rare(
        fastaTokenPath + "output_" + str(i) + ".fasta_aa",
    )
    token.update(df_deepcodon_token)
```

```python
for line, value in token.items():
    df_deepcodon.loc[df_deepcodon["id"] == int(line), "token"] = value[2]
```

```python
df_combined = pd.merge(df_deepcodon, df_geneScript, on="id")
df_combined = pd.merge(df_combined, df_icor, on="id")
df_combined = pd.merge(df_combined, df_codonop, on="id")
df_combined = pd.merge(df_combined, df_deepcodonft[["id", "deepcodonft"]], on="id")
df_combined_out = df_combined[
    ["id", "token", "deepcodon", "deepcodonft", "genescript", "icor", "codonop"]
]
```

```python
def assert_all_codon_rare(df, incolumn, outcolumn):
    filtered_codons = [
        k for k, v in property.items() if v < 0.3 and k not in {"<pad>", "<S>", "<E>"}
    ]
    for line, value in df.iterrows():
        tokens = ""
        seq = value[incolumn]
        for i in range(0, len(seq), 3):
            codon = seq[i : i + 3]
            if codon in filtered_codons:
                tokens += "1"
            else:
                tokens += "0"
        df.loc[line, outcolumn] = tokens
    return df
```

```python
df_combined_out = assert_all_codon_rare(df_combined_out, "deepcodon", "deepcodon_rare")
df_combined_out = assert_all_codon_rare(
    df_combined_out, "deepcodonft", "deepcodonft_rare"
)
df_combined_out = assert_all_codon_rare(
    df_combined_out, "genescript", "genescript_rare"
)
df_combined_out = assert_all_codon_rare(df_combined_out, "icor", "icor_rare")
df_combined_out = assert_all_codon_rare(df_combined_out, "codonop", "codonop_rare")
```

```python
df_combined_out
```

```python
def count_right(df, column):
    for line, value in df.iterrows():
        count = 0
        all_count = 0
        for t, v in zip(value["token"], value[column]):
            if t == "1":
                all_count += 1
                if t == v:
                    count += 1
        df.loc[line, column + "_count"] = count
        df.loc[line, "all_count"] = all_count
    return df
```

```python
df_combined_out = count_right(df_combined_out, "deepcodon_rare")
df_combined_out = count_right(df_combined_out, "deepcodonft_rare")
df_combined_out = count_right(df_combined_out, "genescript_rare")
df_combined_out = count_right(df_combined_out, "icor_rare")
df_combined_out = count_right(df_combined_out, "codonop_rare")
```
